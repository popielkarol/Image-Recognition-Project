```python
import numpy as np
import cv2
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.measure import label, regionprops
import matplotlib.image as mpimg
```

```python
def preprocess_image(image):
    # Stosowanie rozmycia Gaussa do zmniejszenia szumów
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
    return blurred_image
```

```python
def find_dominant_colors(image, num_clusters=1):
    # Konwersja obrazu do tablicy 1-wymiarowej
    pixels = np.float32(image.reshape(-1, 3))
    
    # Użycie algorytmu KMeans do znalezienia dominantnych kolorów
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    kmeans.fit(pixels)
    
    # Pobranie etykiet klastrów oraz ich centroidów
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    
    return centers.astype(int)
```

```python
def create_color_masks(image, centers):
    masks = []
    for center in centers:
        lower = np.maximum(center - 30, 0)
        upper = np.minimum(center + 30, 255)
        mask = cv2.inRange(image, lower, upper)
        masks.append(mask)
    return masks
```

```python
def detect_squares_using_clustering(masks, scale_factor=1):
    # squares okresla informację o położeniu wykrytych kwadratów
    squares = []
    for mask in masks:
        labeled_mask = label(mask == 0)
        for region in regionprops(labeled_mask):
            if region.area >= 25:
                minr, minc, maxr, maxc = region.bbox
                w = (maxc - minc) * scale_factor
                h = (maxr - minr) * scale_factor
                if 0.9 <= w / h <= 1.1:
                    squares.append((minc * scale_factor, minr * scale_factor, w, h))
        plt.imshow(labeled_mask, cmap='gray')  # Opcjonalna wizualizacja
        plt.show()
    return squares
```

```python
# Wygenerowanie pliku z losowym tłem (macierz 500x500x3)
random_image = np.random.randint(1, 50, size=[500, 500, 3], dtype=int)

plt.imshow(random_image, cmap='gray')
plt.colorbar();

print('Type:', type(random_image))
print('dtype:', random_image.dtype)
print('shape:', random_image.shape)
```

```python
#Dodajemy kwadraciki 10x10 o określonych kolorach [R,G,B]
random_image[50:60, 20:30, :] = [255, 59, 59]

random_image[250:260, 200:210, :] = [213, 255, 0]
random_image[410:420, 200:210, :] = [255, 0, 230]

random_image[150:170, 90:110, :] = [76, 0, 47]
random_image[470:480, 200:210, :] = [25, 0, 230]
random_image[23:33, 20:30, :] = [0, 255, 55]
plt.imshow(random_image);
random_image = random_image.astype(np.uint8)
mpimg.imsave('random_dots.png', random_image)
```

```python
# Load and preprocess the image
image_path = 'random_dots.png'
image = imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
preprocessed_image = preprocess_image(image)

# Find dominant colors
# Zmień parametr num_clusters w funkcji find_dominant_colors na 1
dominant_colors = find_dominant_colors(preprocessed_image, num_clusters=1)

# Tworzenie maski dla jednego dominującego koloru
masks = create_color_masks(preprocessed_image, dominant_colors)

# Detekcja kwadratów
detected_squares = detect_squares_using_clustering(masks)


#Zapis obrazka do pliku
random_image = random_image.astype(np.uint8)
mpimg.imsave('random_dots.png', random_image)
```

```python
# Wyświetlanie wyników
for mask in masks:
    plt.imshow(mask, cmap='gray')
    plt.title('Mask for dominant color')
    plt.show()

for (x, y, w, h) in detected_squares:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
plt.title('Detected Squares on Original Image')
plt.show()

print("Detected squares:", detected_squares)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def determine_square_color(region):
    # Obliczenie średniego koloru obszaru
    avg_color = np.mean(region, axis=(0, 1))

    # Lista dostępnych kolorów w formacie BGR
    available_colors = [
        ("Czarny", (0, 0, 0)),
        ("Biały", (255, 255, 255)),
        ("Szary", (128, 128, 128)),
        ("Czerwony", (0, 0, 200)),       
        ("Zielony", (0, 150, 0)),       
        ("Niebieski", (150, 0, 0)),     
        ("Zolty", (0, 150, 150)),      
        ("Fioletowy", (150, 0, 150)), 
        ("Turkusowy", (0, 150, 150)),  
        ("Różowy", (150, 100, 200)),    
        ("Pomaranczowy", (0, 100, 200)), 
        ("Brazowy", (0, 60, 120))      
    ]

    # Wyszukiwanie najbliższego koloru
    min_distance = float('inf')
    closest_color = None
    for color, color_val in available_colors:
        distance = np.linalg.norm(avg_color - color_val)
        if distance < min_distance:
            min_distance = distance
            closest_color = color

    return closest_color

# Wywołanie funkcji dla wykrytych kwadratów
detected_squares_colors = []
for (x, y, w, h) in detected_squares:
    region = preprocessed_image[int(y):int(y+h), int(x):int(x+w)]
    square_color = determine_square_color(region)
    detected_squares_colors.append(square_color)
    # Dodanie napisu na obrazie oryginalnym
    cv2.putText(image, square_color, (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
plt.title('Detected Squares on Original Image')
plt.show()

print("Kolory wykrytych kwadratów:", detected_squares_colors)


```

```python
# Load image
imageObj = cv2.imread('random_dots.png') 
# When we read the image using cv2.imread() method, the image read is in BGR format

# To avoid grid lines
plt.axis("off") 
plt.title("Original Image") 
plt.imshow(cv2.cvtColor(imageObj, cv2.COLOR_BGR2RGB)) 
plt.show() 

# Get BGR data from image
blue_color = cv2.calcHist([imageObj], [0], None, [256], [0, 256]) 
green_color = cv2.calcHist([imageObj], [1], None, [256], [0, 256]) 
red_color = cv2.calcHist([imageObj], [2], None, [256], [0, 256]) 

plt.figure(num=None, figsize=(8, 2))
plt.plot(blue_color, color='b') 
plt.plot(green_color, color='g') 
plt.plot(red_color, color='r') 

# Adding labels to axes
plt.title('Image Histogram') 
plt.xlabel('Intensity') 
plt.ylabel('Pixels') 

plt.show()
```
